## read_csv()

```python
## lendo o arquivo csv
df = pd.read_csv('data/Breast_cancer_data.csv')

## inspecionando as 5 primeiras linhas do dataframe
df.head()
```

```python
## relendo o csv, mas agora pulando as 10 primeiras linhas (skiprows=10) e as últimas 10 (skipfooter=10) e atribuindo nenhum cabeçalho (header=None)
df = pd.read_csv('data/Breast_cancer_data.csv',skiprows=10,skipfooter=10,header=None)
df.head()
```

```python
## salvando o dataframe df em 'data/dados_sem_10.csv' com o separador ';'
df.to_csv('data/dados_sem_10.csv',sep=';')
```
